```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python

# 4.1 Univariate Analysis
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Age Distribution
axes[0,0].hist(df['Age'], bins=30, edgecolor='black', alpha=0.7)
axes[0,0].set_title('Age Distribution')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')

# Purchase Amount Distribution
axes[0,1].hist(df['Purchase Amount (USD)'], bins=30, edgecolor='black', alpha=0.7, color='green')
axes[0,1].set_title('Purchase Amount Distribution')
axes[0,1].set_xlabel('Purchase Amount (USD)')

# Review Rating Distribution
axes[0,2].hist(df['Review Rating'], bins=20, edgecolor='black', alpha=0.7, color='orange')
axes[0,2].set_title('Review Rating Distribution')
axes[0,2].set_xlabel('Review Rating')

# Gender Distribution
gender_counts = df['Gender'].value_counts()
axes[1,0].bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
axes[1,0].set_title('Gender Distribution')
axes[1,0].set_ylabel('Count')

# Category Distribution
top_categories = df['Category'].value_counts().head(10)
axes[1,1].barh(top_categories.index, top_categories.values)
axes[1,1].set_title('Top 10 Categories')
axes[1,1].set_xlabel('Count')

# Season Distribution
season_counts = df['Season'].value_counts()
axes[1,2].bar(season_counts.index, season_counts.values, color=['red', 'blue', 'green', 'orange'])
axes[1,2].set_title('Season Distribution')
axes[1,2].set_ylabel('Count')

plt.tight_layout()
plt.savefig('univariate_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
print("""Key Insights from Plots
 The customer age distribution is broad and fairly uniform, indicating a diverse user base with no dominant age group.
 Purchase amounts are evenly spread across the range, suggesting consistent spending behavior with frequent small to medium transactions.
 Review ratings are mostly concentrated between 3.0 and 4.5, reflecting overall positive customer satisfaction.
 The dataset is male-dominated, indicating an opportunity for targeted gender-based marketing strategies.
 Clothing and Accessories are the top-performing categories, contributing the majority of sales volume.
 Sales are evenly distributed across seasons, showing minimal seasonal dependency and stable year-round demand.
Business Implication
The business demonstrates stable purchasing behavior across demographics and seasons, with strong performance in Clothing and Accessories. Marketing and inventory strategies should prioritize these categories while leveraging data-driven targeting to optimize customer engagement.""")
```

```python
# 4.3 Correlation Analysis
# Select numerical columns
numerical_cols = ['Age', 'Purchase Amount (USD)', 'Review Rating', 'Previous Purchases']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix of Numerical Variables')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['Age'], q=4, labels = labels)
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase frequency days'] = df['Frequency of Purchases'].map(frequency_mapping)
```

```python
df.head(10)
```

```python
df['Age_Group'] = pd.cut(df['Age'], 
                               bins=[0, 25, 35, 50, 65, 100],
                               labels=['18-25', '26-35', '36-50', '51-65', '65+'])

print("\n1. Customer Demographics:")
print(f"   Total Customers: {df['Customer ID'].nunique()}")
print(f"   Average Age: {df['Age'].mean():.1f} years")
print(f"   Gender Distribution:\n{df['Gender'].value_counts(normalize=True).mul(100).round(1)}")

print("\n2. Purchase Behavior:")
print(f"   Average Purchase Amount: ${df['Purchase Amount (USD)'].mean():.2f}")
print(f"   Total Revenue: ${df['Purchase Amount (USD)'].sum():,.2f}")
print(f"   Most Popular Category: {df['Category'].mode()[0]}")
print(f"   Average Previous Purchases: {df['Previous Purchases'].mean():.1f}")

print("\n3. Customer Satisfaction:")
print(f"   Average Review Rating: {df['Review Rating'].mean():.2f}/5")
print(f"   Subscription Rate: {(df['Subscription Status'] == 'Yes').mean()*100:.1f}%")

print("\n4. Marketing Effectiveness:")
discount_rate = (df['Discount Applied'] == 'Yes').mean() * 100
promo_rate = (df['Promo Code Used'] == 'Yes').mean() * 100
print(f"   Discount Usage: {discount_rate:.1f}%")
print(f"   Promo Code Usage: {promo_rate:.1f}%")

```
